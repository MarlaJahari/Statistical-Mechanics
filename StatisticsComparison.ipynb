```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
e = 1.6e-19              
k = 1.38e-23              
u = 0                            #Electric Charge, Boltzmann constant, Chemical potential 
n=int(input("No: of steps:"))    #No: of steps
En = np.linspace (-0.5,0.5,n)    #Range of energy 
```

```python
def mb(T,E):
    return 1/(np.exp(((E-u)*e)/(k*T)))
```

```python
m=int(input("No: of temperatures: "))
T = np.zeros(m)
M = np.zeros([m,n])
for i in range (m) :
    T[i] = float(input("Enter Temperature: ")
    for j in range (n) :
       M[i,j] = mb(T[i],En[j])
       
```

```python
for i in range (m):
    plt.plot(En,M[i,:])
    plt.xlabel('Energy (eV)')
    plt.ylabel('f(E)')
    plt.xlim(-0.3,0.3)
    plt.ylim(0,3)
    plt.title ('Maxwell-Boltzmann Distribution ')
    plt.legend(["{}".format(T[i]) for i in range(m)])
```

```python
def ferD(T,E):
    return 1/((np.exp(((E-u)*e)/(k*T))) + 1)
```

```python
F= np.zeros([m,n])
for i in range (m):
    for j in range(n):
       F[i,j] = ferD(T[i],En[j])
```

```python
for i in range (m):
    plt.plot(En,F[i,:])
    plt.xlabel('Energy (eV)')
    plt.ylabel('f(E)')
    plt.xlim(-0.3,0.3)
    plt.ylim(0,1)
    plt.title ('Fermi-Dirac Distribution ')
    plt.legend(["{}".format(T[i]) for i in range(m)])
```

```python
def BoseE(T,E):
    return 1/((np.exp(((E-u)*e)/(k*T)))-1)
```

```python
B = np.zeros([m,n])
for i in range (m) :
    for j in range (n) :
       B[i,j] = BoseE(T[i],En[j])
```

```python
for i in range (m):
    plt.plot(En,B[i,:])
    plt.xlabel('Energy in eV --->')
    plt.ylabel('f(E) --->')
    plt.xlim(0,0.4)
    plt.ylim(0,3)
    plt.title ('Fermi Dirac Distribution ')
    plt.legend(["{}".format(T[i]) for i in range(m)])
```

```python
for i in range(n):
    B[1,i] = BoseE(2000,En[i])
    F[1,i] = ferD(2000,En[i])
    M[1,i] = mb(2000,En[i])   
```

```python
plt.plot(En,F[1,:])
plt.plot(En,M[1,:])
plt.plot(En,B[1,:])
plt.ylim(0,4)
plt.grid()
plt.legend(["Fermi-Dirac","Maxwell_Boltzmann","Bose-Einstein"])
plt.xlabel('Energy in eV --->')
plt.ylabel('f(E) --->')
```
