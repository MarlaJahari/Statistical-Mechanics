Required Formulae

1) Frequency  based

Rayleigh Jean Law: U(v)dv=8*pi*k*T*v*v/c**3 dv

Planck Law: U(v)dv=8*pi*h*v**3/(c**3 * (exp(h*v/(k*T))-1)) dl
    
2) Wavelength based (l-lambda)

Rayleigh Jean Law: U(l)dl=8*pi*k*T/l**4 dl

Planck Law: U(v)dv=8*pi*h*c/(l**5 * (exp(h*c/(k*l*T))-1)) dl

```python
def u(x):
    return x**3*np.exp(-x)/(np.exp(-x)-1)
```

```python
from scipy import integrate
inf = math.inf
integ = integrate.quad(u,0,inf)

sc =  8*math.pi*k**4/(4*(h**3)*(c**2))*integ[0]
Sc = 2* math.pi**5 *k**4/(15*h**3*c**2)

print("Stefan's constant by quadrature method of integration :",abs(sc))
print("Stefan's constant by formula: ",Sc)
```
