*BOSONS*

$n(E) =\frac{1}{[\exp(\beta(E - μ)) -1 ]}$  where  $β= \frac{1}{kT}$

*Non-relativistic*:
$\frac{dN}{dE} = C_{n} n(E) \sqrt{E}$  where  $C_{n}=(2s+1) \frac{2\pi V (2m)^\frac{3}{2}}{h^3}$

*Relativistic*:
$\frac{dN}{dE} = C_{r} n(E) E^2$  where  $C_{r}= 2s \frac{4 \pi V}{(h^3 c^3)}$

*FERMIONS*

$n(E) =\frac{1}{[\exp(\beta(E - μ)) +1 ]}$

*Non-relativistic*:
$\frac{dN}{dE} = C_{n} n(E) \sqrt{E}$ 

*Relativistic*:
$\frac{dN}{dE} = C_{r} n(E) E^2$ 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
me = 0.511e+6
h = 6.626e-34/1.6e-19
c = 3e+8
k = 1.38e-23/1.6e-19
pi = np.pi
E = np.linspace(1e-5,0.5,1000)
V = 1
e = 1.6e-27
```

```python
def nr(E,T,s,mu):
    beta = 1/(k*T)
    Cn = (2*s+1)*(2*pi*V*(2*me)**1.5)/h**3
    nb = 1/(np.exp(beta*(E-mu))-1) 
    nf = 1/(np.exp(beta*(E-mu))+1)
    if s//1 == s:
        return Cn*(nb)*np.sqrt(E)
    else:
        return Cn*nf*np.sqrt(E)
```

```python
#s = float(input('Enter the spin of particle: '))
mu1 = float(input("Chemical potential energy(keep it negative, boson here) : "))
plt.subplot(1,2,1)
plt.plot(E,nr(E,100,1,mu1))
plt.title("Bosons")

mu2 = float(input("Chemical potential energy(keep it <0.5, fermion here) : "))
plt.subplot(1,2,2)
plt.plot(E,nr(E,100,0.5,mu2))
plt.title("Fermions")

plt.suptitle("Non-Relativistic plots")
```

```python
def rel(E,T,s,mu):
    beta = 1/(k*T)
    Cr = (2*s)*(4*pi*V)/(h**3 * c**3)
    nb = 1/(np.exp(beta*(E-mu))-1)
    nf = 1/(np.exp(beta*(E-mu))+1)
    if s//1 == s:
        return Cr*nb*(E)**2
    else:
        return Cr*nf*(E)**2
```

```python
plt.subplot(1,2,1)
plt.plot(E,rel(E,100,1,mu1))
plt.title("Bosons")

plt.subplot(1,2,2)
plt.plot(E,rel(E,100,0.5,mu2))
plt.title("Fermions")

plt.suptitle("Relativistic plots")
```
