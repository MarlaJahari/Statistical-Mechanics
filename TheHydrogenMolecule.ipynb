```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse.linalg import eigsh
```

```python

hbar=197.3
n=1000
alpha= 1.945
D=4.75
r0=0.7417
mu=(940/2)*1e+6     
r= np.linspace(0.1,1,n)
d=r[1]-r[0]          
```

```python

def pot(r):
  return D*(np.exp(-2*alpha*((r-r0)))-2*np.exp(-alpha*((r-r0))))
```

```python

k=(hbar*hbar)/(2*mu*d*d)
X=[]
Y=[]

for i in range(n-1):
    X.append(k)
for i in range(n):
    Y.append(-2*k)

K=np.diag(X,-1)+np.diag(Y,0)+np.diag(X,1)
```

```python

v= pot(r)
V=np.diag(v,0)
```

```python
Ev,psiv=np.linalg.eigh(-K+V)
```

```python
for i in range(2):
     plt.plot(r,(psiv[:,i]**2))
plt.title('Wave Functions for Vibrational Mode')
plt.grid(True)
plt.xlabel('r')    
plt.ylabel('Wave Function')
plt.legend()
```

```python
print("Ground state energy (eV) :",np.round(Ev[0],3))
```
